```python
#판다스 라이브러리 선언
import pandas as pd
from pandas import Series, DataFrame

myCampus = Series([100,200,300,400,500], 
    index=["2026-07-01","2026-07-02","2026-07-03","2026-07-04","2026-07-05"])

myCampus
#셀을 실행할 때 shift + enter를 누르면 바로 아래에 결과가 출력됨
#슬라이싱
myCampus["2026-07-05"] 
```

```python
#데이터프레임을 로딩
data = { "foreigner":[1,2,3,4,5,6], 
         "sratio":[10,20,30,40,50,60], 
         "org":[100,200,300,400,500,600],
         "sprice":[1000,2000,3000,4000,5000,6000], 
         "private":[1,2,3,4,5,6]
        }

#컬럼을 순서를 지정
frame1 = DataFrame(data=data, 
    columns=["private", "org", "foreigner","sratio","sprice"])
frame1 
```

```python
frame2 = DataFrame(data=data, 
    columns=["private", "org", "foreigner","sratio","sprice"], 
    index=["2026-07-01","2026-07-02","2026-07-03","2026-07-04",
        "2026-07-05","2026-07-06"])
#컬럼을 슬라이싱
frame2["org"] 
```

```python
#특정행: .loc => location약자 
frame2.loc["2026-07-01"]
```
